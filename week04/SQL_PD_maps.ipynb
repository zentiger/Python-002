9. DELETE FROM table1 WHERE id=10;

```python
raw_data = {
        'id': ['1', '2', '3', '4', '5', '10'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung', 'GeeK'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches', 'Times']}

df3 = pd.DataFrame(raw_data, columns=['id','first_name','last_name'])
df3 = df3[df3['id']!='10']
print(df3)
```
